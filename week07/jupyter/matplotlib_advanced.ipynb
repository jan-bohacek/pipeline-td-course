```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)
y = x**2 + 2*x + 2

fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, y*1, color="red", linewidth=0.2)
ax.plot(x, y*2, color="red", linewidth=0.5)
ax.plot(x, y*3, color="red", linewidth=1.0)

ax.plot(x, y*4, color="green", linestyle="--")
ax.plot(x, y*5, color="green", linestyle="-.")
ax.plot(x, y*6, color="green", linestyle=":")

line = ax.plot(x, y*7, color="blue", lw=1.5)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, y**2, x, np.exp(x))
ax.set_title("Numbers")
ax.set_yticks([0,500,1000,1500])

from matplotlib import ticker

ft = ticker.ScalarFormatter(useMathText=True)
ft.set_scientific(True)
ft.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(ft)
ax.yaxis.labelpad = 15
ax.xaxis.labelpad = 10
```

```python
fix, axes = plt.subplots(1,2, figsize=(10,3))

axes[0].plot(x, x**2, x, x**3, lw=2) # lineweight is 2
axes[0].grid(True)
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Fucked-up-ness")

axes[1].plot(x, x**2, x, x**3, lw=3)
axes[1].grid(color="b", alpha=0.5, linestyle="--", lw=0.5)
```

```python
n = np.arange(1,50)
x = np.linspace(-0.5, 1.0, 50)
```

```python
fig, ax = plt.subplots(1,4, figsize=(12,3))

ax[0].scatter(x, x+0.5 * np.random.randn(len(x)))
ax[0].set_title("Scatter")

ax[1].step(n, n**2, lw=1.3)
ax[1].set_title("Step")

ax[2].bar(n, n**2, align="center", width=0.5, alpha=0.7)
ax[2].set_title("Bar")

ax[3].fill_between(x, x**2, x**3, color="green", alpha=0.3)
ax[3].set_title("Fill")
```

# 3D Plotting

```python
from  mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
X,Y = np.meshgrid(np.linspace(0,2*np.pi, 300), np.linspace(0.2*np.pi, 300))
Z = 2*np.cos(X)
```

```python
fig = plt.figure(figsize=(10,5))

ax = fig.add_subplot(1,2,1, projection="3d")
p = ax.plot_surface(X, Y, Z, rstride=5, cstride=5, linewidth=1.0)

ax = fig.add_subplot(1,2,2, projection="3d")
p = ax.plot_surface(X, Y, Z, rstride=10, cstride=10, linewidth=1.0, cmap=matplotlib.cm.viridis, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)

ax.contour(X,Y,Z, zdir="x", offset=-np.pi, cmap="coolwarm")
ax.contour(X,Y,Z, zdir="y", offset=-np.pi, cmap="coolwarm")
ax.contour(X,Y,Z, zdir="z", offset=5*np.pi, cmap="coolwarm")
```
