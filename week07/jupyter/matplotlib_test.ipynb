```python
NUM_CLASSES = 4
NUM_FEATURES = 2

x_blob, y_blob = datasets.make_blobs(n_samples=1000,
                                     n_features=NUM_FEATURES,
                                     centers=NUM_CLASSES,
                                     cluster_std=1.1,
                                     random_state=1213
                                    )
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x_blob[:,0], x_blob[:,1], c=y_blob, cmap=plt.cm.copper);
```
