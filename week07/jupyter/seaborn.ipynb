```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_flights = sns.load_dataset("flights")
```

```python
data_flights.head()
```

```python
sns.histplot(data=data_flights, x="year", y="passengers", hue="year", multiple="stack");
```

```python
data_penguins = sns.load_dataset("penguins")
```

```python
sns.kdeplot(data=data_penguins, x="body_mass_g", hue="species", multiple="stack");
```

```python
sns.displot(data=data_penguins, x="body_mass_g", hue="sex", multiple="stack");
```

# Multiple plots (visualize correlation)

```python
data_tips = sns.load_dataset("tips")
```

```python
data_tips.head()
```

```python
f, axs = plt.subplots(2,3, figsize=(8,4), gridspec_kw=dict(width_ratios=[4,3,2]))
```

```python
f, axs = plt.subplots(1,2, figsize=(8,4), gridspec_kw=dict(width_ratios=[4,3]))
sns.scatterplot(data=data_tips, x="sex", y="tip", hue="sex", ax=axs[0])
sns.histplot(data=data_tips, x="sex", hue="sex", shrink=.5, legend=False)
f.tight_layout()
```
